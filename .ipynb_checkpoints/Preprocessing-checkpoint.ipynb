# Load Data

```python
import pandas as pd
from pathlib import Path
from tqdm import tqdm_notebook as tqdm
```

```python
dataset_dir = '../ExtraSensory.per_uuid_features_labels'
data = pd.DataFrame()
for filename in tqdm(Path(dataset_dir).iterdir()):
    csv = pd.read_csv(filename, index_col=0)
    key = filename.name[:filename.name.find('.')]
    csv = pd.concat([csv], keys=[key], names=['filename'])
    data = csv if data.empty else data.append(csv)
```
